Ride Sharing Data Observations:

1. The number of Urban cities outweigh the number of Suburban or Rural Cities. In the similar lines, number of rides and drivers in Urban are far greater compared to those in the Subarban and Rural categories. Given fewer rides in the rural Cities, the average fare is more compared to the average fares in other groups.
2. Total Fares and Total Rides percentages are higher for Urban cities compared to Suburban and Rural cities. This shows that the fares are proportionate to the rides.
3. Whopping 86% of drivers belong to Urban Cities. This shows that more drivers prefer to drive in Urban Cities.

Based on the above observations, it is evident that the ride sharing market is more prominent in Urban cities, and fairly good in Suburban cities and poor in Rural cities.

```python
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride_combined_data_df = pd.merge(ride_data_df,city_data_df, on="city", how="left")

# Display the data table for preview
city_ride_combined_data_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
city_ride_grp = city_ride_combined_data_df.groupby(["type","city"])
city_ride_grp_df = pd.DataFrame({"Total Rides": city_ride_grp["ride_id"].count()
                                ,"Average Fare": city_ride_grp["fare"].mean()
                                ,"Driver Count": city_ride_grp["driver_count"].mean() 
                                })

# Build the scatter plots for each city types
fig = plt.figure(figsize=(10,4))
ax = plt.subplot(111)
plt.subplots_adjust(right=0.6)

city_ride_grp_df = city_ride_grp_df.reset_index()

city_type_df = city_ride_grp_df.loc[city_ride_grp_df["type"]=='Urban', :]
urban = ax.scatter(city_type_df["Total Rides"], city_type_df["Average Fare"], s=city_type_df["Driver Count"]**1.55\
            , marker='o', c='coral', edgecolors='black', alpha=0.75, linewidths=1, label='Urban')

city_type_df = city_ride_grp_df.loc[city_ride_grp_df["type"]=='Suburban', :]
suburban = ax.scatter(city_type_df["Total Rides"], city_type_df["Average Fare"], s=city_type_df["Driver Count"]**1.55\
            , marker='o', c='skyblue', edgecolors='black', alpha=0.75, linewidths=1, label='Suburban')

city_type_df = city_ride_grp_df.loc[city_ride_grp_df["type"]=='Rural', :]
rural = ax.scatter(city_type_df["Total Rides"], city_type_df["Average Fare"], s=city_type_df["Driver Count"]**1.55\
            , marker='o', c='gold', edgecolors='black', alpha=0.75, linewidths=1, label='Rural')

# Incorporate the other graph properties
ax.set_xlabel("Total Number of Rides (Per City)")
ax.set_ylabel("Average Fare ($)")
ax.set_title("Pyber Ride Sharing Data (2016)")
ax.grid()

# Create a legend
lgnd = ax.legend(handles=[urban,suburban,rural],title="City Types", loc='best')
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
ax.text(42,37,s="Note:\nCircle size correlates with driver count per city."\
         ,horizontalalignment='left',verticalalignment='center'\
        ,clip_on=False,)

# Save Figure
plt.savefig("./Images/PyberRideSharingScatter.png",bbox_inches='tight',dpi=100)
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
city_type_grp = city_ride_combined_data_df.groupby(["type"])
city_type_grp_df = pd.DataFrame({"Total Fares": city_type_grp["fare"].sum()})
city_type_grp_df = city_type_grp_df.reset_index()
city_type_grp_df = city_type_grp_df.sort_values(["type"], ascending=False)

# Build Pie Chart
explode = [0.1,0,0]
colors = ["Coral","Skyblue","Gold"]

fig = plt.figure(2,figsize=(6,3), frameon=True)

plt.pie(city_type_grp_df["Total Fares"], explode=explode, labels=city_type_grp_df["type"]\
        , autopct="%1.1f%%", counterclock=False, startangle=150, shadow=True, colors=colors)
plt.title('% Of Total Fares by City Type')

# Save Figure
plt.savefig("./Images/PyberFaresPie.png",bbox_inches='tight',dpi=100)
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
city_type_grp_df = city_type_grp_df.set_index(["type"])
city_type_grp_df ["Total Rides"] = city_type_grp["ride_id"].count()
city_type_grp_df = city_type_grp_df.reset_index()

# Build Pie Chart
explode = [0.1,0,0]
colors = ["Coral","Skyblue","Gold"]

fig = plt.figure(3,figsize=(6,3), frameon=True)

plt.pie(city_type_grp_df["Total Rides"], explode=explode, labels=city_type_grp_df["type"]\
        , autopct="%1.1f%%", counterclock=False, startangle=150, shadow=True, colors=colors)
plt.title('% Of Total Rides by City Type')

# Save Figure
plt.savefig("./Images/PyberRidesPie.png",bbox_inches='tight',dpi=100)
```

```python
# Show Figure
plt.show();
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
city_type_grp_df = city_type_grp_df.set_index(["type"])
city_type_grp_df ["Total Drivers"] = city_type_grp["driver_count"].sum()
city_type_grp_df = city_type_grp_df.reset_index()

# Build Pie Chart
explode = [0.1,0,0]
colors = ["Coral","Skyblue","Gold"]

fig = plt.figure(4,figsize=(6,3), frameon=True)

plt.pie(city_type_grp_df["Total Drivers"], explode=explode, labels=city_type_grp_df["type"]\
        , autopct="%1.1f%%", counterclock=False, startangle=150, shadow=True, colors=colors)
plt.title('% Of Total Drivers by City Type')

# Save Figure
plt.savefig("./Images/PyberDriversPie.png",bbox_inches='tight',dpi=100)
```

```python
# Show Figure
plt.show()
```
